```python
cols = ['Rating','Review Text', 'Age', 'Recommended IND']
df = df[cols]
df = df.dropna()
print(df.shape)
df.head()
```

```python
#seleksi data dengan rating 1,2,5
df = df[df.Rating.isin([1,2,4,5])]
print(df.shape)
df.head()
```

```python
df['sentiment'] = ['positif' if x==5 or x==4 else 'negatif' for x in df.Rating]
print(df.shape)
df.head()
```

```python
# Menghapus data yang duplikat
df.drop_duplicates(inplace=True)
```

```python
df.dtypes
```

```python
#downsampling
# num_sample = len(df[df.sentiment=='negatif'])
num_sample = 2370
ds_positif_label = df[df.sentiment=='positif'].sample(n=num_sample, random_state=1)
ds_positif_label.head()
```

```python
df = pd.concat([ds_positif_label,df[df.sentiment=='negatif'].iloc[:num_sample]],axis=0)
df.drop('Rating',axis=1,inplace=True)
print(df.shape)
df.head()
```

```python
df.rename(columns=lambda x: x.replace(' ', '_'), inplace=True)

print(df.head())
```

##Casefolding

```python
df['corpus'] = df.Review_Text.str.lower()
print(df.shape)
df.head()
```

##Punctuation

```python
df['corpus'] = df.corpus.str.translate(str.maketrans('', '', string.punctuation))
print(df.shape)
df.head()
```

##Repetition Character

```python
def repetition_char(txt):
  sub_str = re.findall(r"((\w)\2{2,})",txt)
  for x in sub_str:
    txt = txt.replace(x[0],x[1])
  return txt
```

```python
df['corpus'] = df.corpus.apply(repetition_char)
df = df.reset_index(drop=True)
print(df.shape)
df.head()
```

```python
df.sentiment.value_counts()
```

##Stemming

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()

df['corpus'] = df.corpus.progress_apply(lambda x : stemmer.stem(str(x)))
print(df.shape)
df.head()
```

```python
#save to csv
df.to_csv(pth+'clean_data.csv',index=False)
```
